```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('iris (1).csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

```python
#checking for null values
data.isnull().sum()
```

```python
#splitting dependent and independent variables
x = data.iloc[:, 0:4]
y = data.iloc[:, -1]
print(x)
print(y)
```

```python
#splitting the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=4)
```

```python
#Using gini index
dt=DecisionTreeClassifier(criterion="gini", random_state=4)
model=dt.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(30,15))
dot_data = tree.plot_tree(dt,feature_names=['Sepal_Length (cm)','Sepal_Width (cm)','Petal_Length (cm)','Petal_Width (cm)'],class_names=['Iris-Setosa', 'Iris-Versicolor', 'Iris-Virginica'],
filled=True)
```

```python
#accuracy and classification report
y_pred=model.predict(x_test)
print("\nConfusion Matrix : \n",confusion_matrix(y_test,y_pred))
print("\nAccuracy : ",accuracy_score(y_pred,y_test))
print("\nClassification report : \n",classification_report(y_pred,y_test))
```

```python
#Using entropy
dt=DecisionTreeClassifier(criterion="entropy", random_state=4)
model=dt.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(30,15))
dot_data = tree.plot_tree(dt,feature_names=['Sepal_Length (cm)','Sepal_Width (cm)','Petal_Length (cm)','Petal_Width (cm)'],class_names=['Iris-Setosa', 'Iris-Versicolor', 'Iris-Virginica'],
filled=True)
```

```python
#accuracy and classification report
y_pred=model.predict(x_test)
print("\nConfusion Matrix : \n",confusion_matrix(y_test,y_pred))
print("\nAccuracy : ",accuracy_score(y_pred,y_test))
print("\nClassification report : \n",classification_report(y_pred,y_test))
```
